```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title("A Simple Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

```

# import and read the csv file *gas_prices*.csv

```python


gas = pd.read_csv("gas_prices.csv")
gas

#  np.read_csv("gas_prices.csv")
# df = pd.read_csv("gas_prices.csv")
# print(df)
# \read file


# create a simple graph
plt.plot(gas.Year,gas.USA)
plt.show()

```

apply simple functions

```python

plt.plot(gas.Year,gas.USA)
plt.plot(gas.Year,gas.Canada)
plt.plot(gas.Year,gas["South Korea"])
plt.show()


plt.plot(gas.Year, gas.USA, 'r-', marker='o')
plt.plot(gas.Year,gas.Canada)
plt.plot(gas.Year,gas["South Korea"])
plt.show()

```

# legend function uses label in the plot() function

```python


plt.plot(gas.Year, gas.USA, 'r-', marker='o',label='USA')
plt.plot(gas.Year,gas.Canada,'b-',marker='o',label='Canada')
plt.plot(gas.Year,gas["South Korea"],'g-',marker='o',label='South Korea')
plt.title('Gas Prices over time')
plt.xticks(gas.Year)
plt.xlabel('years')
plt.ylabel('price')
plt.legend()
plt.show()
```

# xticks() or yticks() have different ways


```python

plt.plot(gas.Year, gas.USA, 'r-', marker='o',label='USA')
plt.plot(gas.Year,gas.Canada,'b-',marker='o',label='Canada')
plt.plot(gas.Year,gas["South Korea"],'g-',marker='o',label='South Korea')
plt.title('Gas Prices over time')
plt.xticks(gas.Year[::10])
# ####plt.xticks(gas.Year)
plt.xlabel('years')
plt.ylabel('price')
plt.legend()
```

# figure(figsize=(x,y)) function implimentation and also use xticks and yticks

```python

plt.figure(figsize=(8,5))
plt.plot(gas.Year, gas.USA, 'r-', marker='o',label='USA')
plt.plot(gas.Year,gas.Canada,'b-',marker='o',label='Canada')
plt.plot(gas.Year,gas["South Korea"],'g-',marker='o',label='South Korea')
plt.title('Gas Prices over time')
# ####plt.xticks(gas.Year)
plt.xticks(gas.Year[::3])
# plt.yticks(gas.USA)
plt.xlabel('years')
plt.ylabel('price')
plt.xlim([1990,2009])
plt.legend()
plt.show()
```
